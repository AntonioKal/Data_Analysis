# Απολογισμός Τηλεοπτικών σειρών του Αρχείου της ΕΡΤ

## *Γραφήματα για τις διασωθείσες και μη τηλεοπτικές σειρές της περιόδου 1969 - 1989*

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sbn

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel('datasets/myProjects/Archivio_ERT.xlsx')
```

```python
df.groupby('Format').sum()['Episodes'].plot(kind='pie',shadow=False, title = 'Συνολικές Τηλεοπτικές σειρές της Κρατικής TV ανά φορμάτ', autopct = '%1.1f%%', figsize =(10,8), explode = [0, 0, 0.2, 0])
```

```python
plt.figure(figsize = (12,4))
sbn.barplot(x = 'YearStart', y = 'Episodes', estimator = 'sum', hue= 'Format', errorbar=('ci', 0), data =df[df['Format']!='Άγνωστη μορφή'])
plt.title('Παραγωγή τηλεοπτικών επεισοδίων ανά έτος και φορμάτ')
```

```python
plt.figure(figsize =(9,4))
sbn.barplot(x='Channel', y = 'Deleted', estimator='sum', errorbar=('ci', 0), data=df)
plt.title('Διαγραμμένες Τηλεοπτικές σειρές ανά κανάλι')
```

```python
plt.figure(figsize =(9,4))
sbn.barplot(x= 'Format', y = 'Deleted', estimator = 'sum', errorbar=('ci', 0), data=df[df['Format']!='Άγνωστη μορφή'])
plt.title('Διαγραμμένες Τηλεοπτικές σειρές ανά φορμάτ')
```

```python
plt.figure(figsize=(9,4))
sbn.barplot(x= 'Format', y = 'Deleted', hue = 'Channel', estimator = 'sum', errorbar=('ci', 0), data=df)
plt.title('Διαγραμμένες Τηλεοπτικές σειρές ανά φορμάτ και κανάλι')
```

```python
plt.figure(figsize = (9,4))
sbn.barplot(x = 'YearDeleted', y = 'Deleted',  estimator = 'sum', errorbar=('ci', 0), data =df)
plt.title('Διαγραμμένες Τηλεοπτικές σειρές ανά έτος')
```

```python
plt.figure(figsize = (9,4))
sbn.barplot(x = 'YearDeleted', y = 'Episodes',  estimator = 'sum', errorbar=('ci', 0), data =df)
plt.title('Διαγραμμένα επεισόδια ανά έτος')
```

```python
plt.figure(figsize =(9,4))
sbn.barplot(x = 'Channel', y='Episodes', estimator = 'sum', hue='Deleted', errorbar=('ci', 0),  data=df)
plt.title('Συνολική επισκόπηση τηλεοπτικών επεισοδίων, διασωθέντων και μη')
```

```python
df.groupby(['Format', 'Channel']).count()['Title'].plot(kind='pie',shadow=False, title = 'Αριθμός τηλεοπτικών σειρών ανά κανάλι και μέσο προβολής', autopct = '%1.1f%%', figsize =(10,8))
```

```python
df.loc[df['Channel'] == 'ΥΕΝΕΔ', 'Channel_Actual'] = 'ΕΡΤ2'
df.loc[df['Channel'] == 'ΕΡΤ2', 'Channel_Actual'] = 'ΕΡΤ2'
df.loc[df['Channel'] == 'ΕΤ2', 'Channel_Actual'] = 'ΕΡΤ2'
df.loc[df['Channel'] == 'ΕΙΡΤ', 'Channel_Actual'] = 'ΕΡΤ1'
df.loc[df['Channel'] == 'ΕΡΤ', 'Channel_Actual'] = 'ΕΡΤ1'
df.loc[df['Channel'] == 'ΕΤ1', 'Channel_Actual'] = 'ΕΡΤ1'
```

```python
df.groupby('Channel_Actual').count()['Title'].plot(kind='pie', title = 'Αριθμός σειρών σε ΕΡΤ1 και ΕΡΤ2 συνολικά',shadow=False, explode = [0,0.2], autopct = '%1.1f%%')
```
