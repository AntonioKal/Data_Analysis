```python
import pandas as pd

pd.set_option('display.max_rows', None)
```

```python
df = pd.read_excel('datasets/myProjects/Cretan_heroes.xlsx')
```

```python
import seaborn as sbn
import matplotlib.pyplot as plt
```

```python
MALEVIZI = df.loc[(df['REGION'] == 'ΜΑΛΕΒΙΖΙΟΥ') & (df['CLASS']!='NONE')]
PEDIADA =  df.loc[(df['REGION'] == 'ΠΕΔΙΑΔΟΣ') & (df['CLASS']!='NONE')]
```

```python
plt.figure(figsize=(13,5))
sbn.barplot(x = 'CITY', y='NUM', hue= 'CLASS', estimator = 'sum', data=MALEVIZI)
plt.title('Αγωνιστές Μαλεβιζίου')
```

```python
plt.figure(figsize=(12,5))
sbn.barplot(x='CLASS', y = 'NUM', estimator = 'sum', hue = 'REGION', data = df.loc[df['CLASS']!='NONE'])
plt.title('Τάξεις αγωνιστών ανά επαρχία')
```

```python
plt.figure(figsize=(13,5))
sbn.countplot(x = 'REGION', hue= 'CLASS', data=df.loc[df['CLASS']!='NONE'])
plt.title('Αγωνιστές όλων των επαρχιών ανά τάξη κατάταξης')
```

```python
Rifiutati = df.loc[df['CLASS'] == 'NONE']
```

```python
plt.figure(figsize=(14,5))
sbn.barplot(x = 'REGION', y = 'NUM', estimator = 'sum', data=Rifiutati)
plt.title('Απορριπτέες αιτήσεις αγωνιστών ανά επαρχία')
```

```python
plt.figure(figsize=(13,5))
sbn.countplot(x = 'REGION', data=df)
plt.title('Αγωνιστές όλων των επαρχιών')
```
