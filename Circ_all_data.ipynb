# Απολογισμός δανεισμού και εγγραφής νέων μελών Τοπικής Βιβλιοθήκης

**Χρονική περίοδος 2016-2022**

```python
import pandas as pd
```

```python
df = pd.read_excel("datasets/myProjects/Circ_all_data.xlsx")
```

## Μέσος όρος δανεισμών ανά μήνα

```python
all_months_circ = df.iloc[:, [1,3,5,7,9,11, 13]].mean(axis=1)
```

```python
all_months_circ.plot(kind='bar', xlabel='Μήνας', ylabel='Δανεισμοί', title='Μέσος όρο δανεισμών ανά μήνα από 2016-2022' )
```

## Μήνας με τους περισσότερους δανεισμούς ανά μέσο όρο

```python
July_circ = df.iloc[6, [1,3,5,7,9,11, 13]].mean()
```

## Εγγραφή νέων μελών

**Μέσος όρος εγγραφών ανά μήνα για όλα τα έτη**

```python
patrons_all_months = df.iloc[:, [2,4,6,8,10,12, 14]].mean(axis=1)
```

```python
patrons_all_months.plot(kind= 'bar',xlabel='Μήνας', ylabel = 'Εγγραφές', title='Μέσος όρος εγγραφής νέων μελών από 2016-2022')
```

## Δανεισμοί ανά μήνα

```python
somma_per_mese = df.iloc[:, [1,3,5,7,9,11, 13]].sum(axis=1)
```

```python
somma_per_mese.plot(kind='line', title='Συνολικός αριθμός δανεισμών ανά μήνα από 2016-2022', xlabel='Μήνες', ylabel='Δανεισμοί')
```

## Δανεισμοί ανά έτος


```python
somma_per_anno = df.iloc[:, [1,3,5,7,9,11, 13]].sum()
```

```python
somma_per_anno.plot(kind='barh', title='Συνολικός αριθμός δανεισμών ανά έτος από 2016-2022', xlabel='Έτη', ylabel='Δανεισμοί')
```
