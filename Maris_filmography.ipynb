```python
import pandas as pd

import seaborn as sbn
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel('datasets/myProjects/Maris_filmography.xlsx')
```

```python
plt.figure(figsize=(10,8))
sbn.barplot(x = 'media', y = 'episodes',  estimator = 'sum', palette = 'dark', errwidth =0, data = df)
plt.title('Αριθμός επεισοδίων ανά μέσο μετάδοσης')
```

```python
df.groupby('year').sum()['episodes'].plot(kind = 'line', title='Αριθμός επεισοδίων ανά έτος')
```

```python
df.groupby('serie_title').sum()['episodes'].plot(kind='pie',shadow=False, autopct = '%1.1f%%', figsize =(10,8), explode = [0.2, 0, 0, 0, 0, 0.2, 0, 0, 0])
```
