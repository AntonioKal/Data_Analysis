```python
import pandas as pd

import seaborn as sbn
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel('datasets/myProjects/Maris_filmography.xlsx')
```

```python
df.groupby('year').sum()['episodes'].plot(kind = 'line', title='Αριθμός επεισοδίων ανά έτος')
```

```python
df.groupby('media').sum()['episodes'].plot(kind='barh', title='Αριθμός διασκευασμένων επεισοδίων', xlabel='Κανάλι', ylabel='Αριθμός')
```

```python
plt.figure(figsize=(10,6))
sbn.barplot(x ='episodes', y = 'serie_title', hue='media',  estimator='sum', errorbar=('ci', 0), data=df)
plt.title('Αριθμός επεισοδίων ανά σειρά και κανάλι')
```
